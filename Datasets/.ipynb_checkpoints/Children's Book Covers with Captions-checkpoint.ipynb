```python
from PIL import Image
from IPython.display import display
import os
import json

# Function to read content of jsonl file (metadata.jsonl)
def read_metadata_file(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            json_object = json.loads(line)
            data.append(json_object)
    return data

# Function to display the first n images
def display_images(data, n):
    for i in range(min(n, len(data))):
        image_filename = data[i]['file_name']
        image_caption = data[i]['text']
        image_path = os.path.join('childrens-books', image_filename)
        # Print image caption
        print(image_caption)
        # Open and display the image
        image = Image.open(image_path)
        display(image)

# Calling functions
file_path = os.path.join('childrens-books', 'metadata.jsonl')
jsonl_data = read_metadata_file(file_path)
display_images(jsonl_data, n=5)  # Replace 5 with the desired number of images to display
```
