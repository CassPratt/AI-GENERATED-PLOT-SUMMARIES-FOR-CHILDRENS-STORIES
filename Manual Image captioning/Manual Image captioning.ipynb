# Manual image captioning

```python
# import required modules
import os

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import csv
```

```python
# Images directory
directory = '../Datasets/childrens-books'
```

```python
# Create list of files in directory
dir_list = os.listdir(directory)
```

```python
len(dir_list)
```

```python
# Function to split a list a in n subsets
def split(a, n):
    k, m = divmod(len(a), n)
    return (a[i*k+min(i, m):(i+1)*k+min(i+1, m)] for i in range(n))
```

```python
# Split list of images in 171 subsets of size 10 (index 0 to 170)
split_list = list(split(dir_list,171))
```

```python
# iterate over files in
# that directory
def enter_caption(list_subset):
    for filename in split_list[list_subset]:
        f = os.path.join(directory, filename)
        # checking if it is a file
        if os.path.isfile(f):
            imgplot = plt.imshow(mpimg.imread(f))
            plt.ion()
            plt.show()
            print(f)
            img_desc = input('Enter image description: ')
            img_dict[filename] = img_desc
            plt.close()
```

```python
# Dictionary to store images captions
img_dict = {}

# Select subset to caption
# Already captioned: 0 to 159
list_subset = 160

# flag to iterate subsets
keep_captioning = True

while keep_captioning:
    enter_caption(list_subset)
    input_resp = input('Continue with next set? (y/n): ')
    if input_resp in 'yY':
        list_subset += 1
    else:
        keep_captioning = False

print('Last subset index was', list_subset)
```

```python
# Open the CSV file in "append" mode
with open('childrens-books-captions.csv', 'a', newline='') as f:
    # Create a dictionary writer with the dict keys as column fieldnames
    writer = csv.writer(f)
    # Append single row to CSV
    for row in img_dict.items():
        writer.writerow(row)
```
